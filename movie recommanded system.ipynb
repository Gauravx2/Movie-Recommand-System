```python
import numpy as np
import pandas as pd
from ipywidgets import interact
```

```python
movies= pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies=movies.merge(credits, on='title')
```

```python
movies.head()
```

```python
#genres
#id
#keywords
#title
#overview
#cast
#crew
movies=movies[['genres','id','keywords','title','overview','cast','crew']]
```

```python
movies.head()
```

```python
movies['genres'][0]
```

```python
movies.dropna(inplace=True)
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        
    return L
```

```python
import ast
ast.literal_eval(movies['genres'][0])
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python

def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
        
    return L
```

```python
movies.head()
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def convert_job(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
        
    return L
```

```python
movies['crew']=movies['crew'].apply(convert_job)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['cast']+movies['crew']+movies['keywords']+movies['genres']+movies['overview']
```

```python
movies.head()
```

```python
new_df=movies[['id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Stuart Little 2')
```

```python
@interact
def movies(Movie=sorted(list(new_df['title']))):
    recommend(Movie)
```
